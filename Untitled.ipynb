```python
import pandas as pd
import matplotlib.pyplot as plt

class RetailAnalytics:
    def __init__(self, file_path):
        self.file_path = file_path
        self.df = None

    # 1. Data Loading
    def load_data(self):
        try:
            self.df = pd.read_csv(self.file_path)
            print("Data loaded successfully")
        except FileNotFoundError:
            print("File not found")
        except Exception as e:
            print("Error:", e)

    # 2. Data Cleaning
    def clean_data(self):
        self.df['Customer'] = self.df['Customer'].fillna('Unknown')
        self.df['City'] = self.df['City'].str.strip().str.title()
        self.df['Date'] = pd.to_datetime(self.df['Date'], errors='coerce')
        self.df['Price'] = pd.to_numeric(self.df['Price'], errors='coerce')

        # Remove invalid rows
        self.df = self.df.dropna(subset=['Date', 'Price', 'Quantity'])
        print("Data cleaned successfully")

    # 3. Feature Engineering
    def feature_engineering(self):
        self.df['Revenue'] = self.df['Quantity'] * self.df['Price']
        print("Feature engineering completed")

    # 4. Data Analysis (EDA)
    def analyze_data(self):
        self.total_revenue = self.df['Revenue'].sum()

        self.top_products = self.df.groupby('Product')['Revenue'].sum().sort_values(ascending=False).head(5)

        self.city_sales = self.df.groupby('City')['Revenue'].sum().sort_values(ascending=False)

        self.df['Month'] = self.df['Date'].dt.to_period('M')
        self.monthly_sales = self.df.groupby('Month')['Revenue'].sum()

        self.category_perf = self.df.groupby('Category')['Revenue'].sum()

        # Collections
        self.customer_spending = self.df.groupby('Customer')['Revenue'].sum().to_dict()
        self.unique_cities = set(self.df['City'])

        self.top_customers = self.df.groupby('Customer')['Revenue'].sum().sort_values(ascending=False).head(5)

        print("Analysis completed")

    # 5. Data Visualization
    def create_charts(self):
        # Top 5 Products
        plt.figure()
        self.top_products.plot(kind='bar')
        plt.title('Top 5 Products')
        plt.xlabel('Product')
        plt.ylabel('Revenue')
        plt.savefig('top_products.png')
        plt.close()

        # City-wise Sales
        plt.figure()
        self.city_sales.head(10).plot(kind='bar')
        plt.title('City-wise Sales')
        plt.xlabel('City')
        plt.ylabel('Revenue')
        plt.savefig('city_sales.png')
        plt.close()

        # Monthly Sales Trend
        plt.figure()
        self.monthly_sales.plot(kind='line')
        plt.title('Monthly Sales Trend')
        plt.xlabel('Month')
        plt.ylabel('Revenue')
        plt.savefig('monthly_trend.png')
        plt.close()

        # Category Contribution
        plt.figure()
        self.category_perf.plot(kind='pie', autopct='%1.1f%%')
        plt.title('Category Contribution')
        plt.ylabel('')
        plt.savefig('category_pie.png')
        plt.close()

        print("Charts created successfully")

    # 6. Report Generation
    def generate_report(self):
        top_product = self.top_products.idxmax()
        top_city = self.city_sales.idxmax()
        top_category = self.category_perf.idxmax()

        report = f"""
Total Revenue: {self.total_revenue}

Top Product: {top_product}
Top City: {top_city}
Top Category: {top_category}

Top 5 Customers:
{self.top_customers}

Unique Cities:
{list(self.unique_cities)}
"""

        with open('final_report.txt', 'w') as f:
            f.write(report)

        print("Report generated successfully")


# Main Program
if __name__ == "__main__":
    file_path = "raw_retail_dataset_500.csv"

    project = RetailAnalytics(file_path)

    project.load_data()
    project.clean_data()
    project.feature_engineering()
    project.analyze_data()
    project.create_charts()
    project.generate_report()

    print("Project completed successfully!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load Data
def load_data(file_path):
    try:
        df = pd.read_csv(file_path)
        print("✅ Data loaded successfully")
        return df
    except:
        print("❌ Error loading file")
        return None

# 2. Clean Data
def clean_data(df):
    df['Customer'] = df['Customer'].fillna('Unknown')
    df['City'] = df['City'].str.strip().str.title()
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

    df = df.dropna(subset=['Date', 'Price', 'Quantity'])
    print("✅ Data cleaned")
    return df

# 3. Feature Engineering
def add_features(df):
    df['Revenue'] = df['Quantity'] * df['Price']
    print("✅ Revenue column added")
    return df

# 4. Analysis
def analyze(df):
    total_revenue = df['Revenue'].sum()

    top_products = df.groupby('Product')['Revenue'].sum().nlargest(5)
    city_sales = df.groupby('City')['Revenue'].sum().nlargest(10)

    df['Month'] = df['Date'].dt.to_period('M')
    monthly_sales = df.groupby('Month')['Revenue'].sum()

    category_perf = df.groupby('Category')['Revenue'].sum()

    top_customers = df.groupby('Customer')['Revenue'].sum().nlargest(5)

    print("✅ Analysis done")

    return total_revenue, top_products, city_sales, monthly_sales, category_perf, top_customers

# 5. Visualization
def create_charts(top_products, city_sales, monthly_sales, category_perf):
    
    top_products.plot(kind='bar', title='Top Products')
    plt.savefig('top_products.png')
    plt.close()

    city_sales.plot(kind='bar', title='City Sales')
    plt.savefig('city_sales.png')
    plt.close()

    monthly_sales.plot(kind='line', title='Monthly Sales')
    plt.savefig('monthly_sales.png')
    plt.close()

    category_perf.plot(kind='pie', autopct='%1.1f%%', title='Category Share')
    plt.savefig('category_pie.png')
    plt.close()

    print("✅ Charts saved")

# 6. Report
def generate_report(total_revenue, top_products, city_sales, category_perf, top_customers):
    
    report = f"""
Total Revenue: {total_revenue}

Top Product: {top_products.idxmax()}
Top City: {city_sales.idxmax()}
Top Category: {category_perf.idxmax()}

Top Customers:
{top_customers}
"""

    with open('report.txt', 'w') as f:
        f.write(report)

    print("✅ Report generated")

# MAIN PROGRAM
file_path = "raw_retail_dataset_500.csv"

df = load_data(file_path)

if df is not None:
    df = clean_data(df)
    df = add_features(df)

    total_revenue, top_products, city_sales, monthly_sales, category_perf, top_customers = analyze(df)

    create_charts(top_products, city_sales, monthly_sales, category_perf)

    generate_report(total_revenue, top_products, city_sales, category_perf, top_customers)

    print("🎉 Project Completed!")
```
